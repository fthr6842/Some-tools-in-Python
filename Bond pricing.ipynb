## bond pricing

```python
def bond_pricing (prin, c, r, k):
    list0 = [round(c/(1+r)**(i+1), 2) for i in range(0, k)]
    list0.append(round(prin/(1+r)**(k), 2))
    return round(sum(list0), 2)
```

```python
for i in range(1, 5+1):
    print("Year ", i, ": ", bond_pricing(1000, 100, 0.05, i))
```

## Example: plot for premium

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([i for i in range(1, 100+1)], [bond_pricing(1000, 100, 0.05, i) for i in range(1, 100+1)])
plt.xlabel("Time to maturity")
plt.ylabel("Price")
```
